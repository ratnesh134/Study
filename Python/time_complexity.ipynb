### Efficiency in programming ?
Time Complexity and Space Complexity

#### Time Complexity

Techniques to measure time efficiency

1. Measuring time (Not a standard method to measure time complexity) Depends upon machine configuration

2. Counting Operations - Assumption in this technique
mathematical operations, comparisons, assignment and accessing objects in memory takes constant time.

Then we count the number of operations executed as function of size of input

Advantage - Now time for different algorithms is same across different machines.

Problems with this technique - Time varies if implementation changes
No clear definition of which operation to count

(Not used in Industry)

3. Orders of Growth (Used in industry) -

BIG O Notations

Types of Orders of Growth

1. Constant Time - example ---> searching an element in a list/ Array . the time taken would be same for a given index , if the size increases also. O(1)

2.  Linear Time --> Searching a database (Assuming the worst case scenario) , if database size increases , the time will also increase . O(n). Linear Search


3. Quadratic Time --> Example is nested loops. O(n^2)

4. Lograthmic time --> O(log(n)) . Example Binary Search

5. nlog(n) --> Example Sorting algorithm

6. Exponential Time --> O(2^n) . Dangerous/ Bad time complexity


```python
## Recursion

def fib(n):
    if n==1 or n==0:
        return 1
    else:
        return fib(n-1) + fib(n-2)

print(fib(5))
```

### Tiime complexity for the previous code 
During recursion we have to calculate number of function calls. therefore here the time complexity for the above program would be O(2^n) , exponential time complexity.
