# Celcius To Farenheit

```python
cel = float(input("Enter the temperature in celcius"))

far = ((9*cel)/5) + 32

print(far)
```

# Euclidean Distance between 2 points

```python
import math
try:
            
    x1,y1 = map(float,input("Enter the first coordinate points for x1 and y1").split())
    x2,y2 = map(float,input("Enter the second coordinate points for x2 and y2").split())
    
    ed = math.sqrt((x2-x1)**2 + (y2-y1)**2)
    
    print(f"The Euclidean Distance is {ed}")
except ValueError:
    print("Enter valid values")
```

```python
# Sum of squares  n natural numbers

n = int(input("Enter the n value"))
sum  = 0
for i in range(1,n+1):
    sum = sum+i

print(f"The sum of {n} natural numbers is {sum}")
```
