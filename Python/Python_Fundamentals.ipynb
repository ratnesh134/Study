# Variables
Variables are containers to hold data types

```python
# Static Vs Dynamic Typing
# Python supports Dynamic Typing
a = 5
# Here python interpreter automatically detects the data type
```

```python
# Dynamic Biding
# Can hold multiple data type
a = 5
print(a)
a = "Ratnesh"
print(a)
```

Identifiers --> Cannot start the variable name with a digit
