# Tuples

A tuple in Python is similar to a list . The difference between the two is that we cannot change the elements of a tuple once it is assigned whereas we can change the elements of a list.

In short, tuple is an immutable list . A tuple cannot be changed in any way once it is created.

Characteristics

1. Ordered
2. Unchangeable
3. Allows duplicates

```python
l = (1,2,2,3)

print(l)
```

```python
type(l)
```

## Creating tuple

```python
# empty tuple

t1 = ()
print(t1)

print(type(t1))
```

```python
# single item tuple

t2 = ("hello")
print(type(t2))
```

```python
t2 = (1)
print(t2)
print(type(t2))
```

```python
# real single item tuple, put a comma after the element

t2 = ("hello",)
print(type(t2))
```

## Accessing items

1. Indexing
2. Slicing

Similar like list

```python
t3 = (1,2,3,4,5)
t4 = ("Hello",23,True,(4,5))

print(t3[3])

print(t4[::-1])
```

```python
print(t4[-1][0])
```

## Editing in tuples

tuples are immutable , so editing is not possible at all.

```python
t3 = (1,2,3,4)
t3[2] = 90

print(t3)
```

## Deleting tuple

we can delete an entire tuple but not an individual item in tuple because then it would mean doing changes in tuple

```python
t3 = (1,2,4,5)


```

```python
t3
```

```python
del t3
```

```python
t3
```

## Operation in tuples

1. Addition and Multiplication
2. Membership
3. Iteration

```python
t1 = (1,2,3,4,5)
t2 = (5,6,7,8,9)

print(t1 + t2)
```

```python
print(t1 * t2)
```
