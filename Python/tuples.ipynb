# Tuples

A tuple in Python is similar to a list . The difference between the two is that we cannot change the elements of a tuple once it is assigned whereas we can change the elements of a list.

In short, tuple is an immutable list . A tuple cannot be changed in any way once it is created.

Characteristics

1. Ordered
2. Unchangeable
3. Allows duplicates

```python
l = (1,2,2,3)

print(l)
```

```python
type(l)
```

## Creating tuple

```python
# empty tuple

t1 = ()
print(t1)

print(type(t1))
```

```python
# single item tuple

t2 = ("hello")
print(type(t2))
```

```python
t2 = (1)
print(t2)
print(type(t2))
```

```python
# real single item tuple, put a comma after the element

t2 = ("hello",)
print(type(t2))
```
