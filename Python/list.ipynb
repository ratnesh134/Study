# Python List


## What is Lists ?

list is a data type where you can store multiple items under 1 name. list act like a dynamic arrays which means you can add more items on the fly.

## Array VS List

Fixed VS Dynamic Size

Convenience (Homogeneous) --> Hetrogeneous

Speed of execution for Python list is slower than Array

Memory is dynamic in Python list and Array has a fixed memory
 



# Memory storage in List



```python
# Proof of the above 

l = [1,2,3]

print(id(l))
print(id(l[0]))
print(id(l[1]))
print(id(l[2]))
print(id(1))
print(id(2))
print(id(3))
```

Benefits of storing addresses

We are not storing exact value , we can store values of multiple data type.  Since we are storing address of the memory location, these types of arrays are called referential arrays.

But it takes a lot of spaces and time (downside).

## Creating a list

```python
# Empty list
l = []
print(l)
```

```python
# 1 dim 
l = [1,2,3,4,5]
print(l)
```

```python
# 2 dim list 
l = [1,2,3,[4,5]]
print(l)
print(type(l))
```

```python
# 3 dim list
l = [[[1,2],[3,4]],[[5,6],[7,8]]]
print(l)
```

## Accessing items from a list

```python
# Indexing
l = [1,2,3,4,5]
print(l[3])

# 2D list
l = [1,2,3,[4,5]]
print(l[3][1])
print(l[0])
```

```python
# Slicing

l = [1,2,3,4,5,6,7,8]
print(l[0:3])
print(l[-3:])
print(l[::-1]) # to reverse the list
```

## Adding items to the list

```python
# append
# adds only one item in the list
l = [1,2,3,4,5]
l.append(6)
print(l)
```

```python
# extend
# can add multiple items in the list
l = [1,2,3,4,5]
l.extend([6,7,8])
print(l)
```

```python
l = [1,2,3,4,5]
l.extend('delhi')
print(l)
```

## Insertion

```python
l = [1,2,3,4,5]
#insert to insert an element at a desired index
l.insert(1,100)

# l.insert(index,element)
print(l)
```

## Editing items in list

```python
l = [1,2,3,4,5,6]
# editing with indexing
l[2] = 500
print(l)

# editing with slicing
l[2:5] = [300,400,500,600,9000,78]
print(l)
```

## Deleting item from a list

```python
# del
# deleting the entire list 
l = [1,2,3,4]
print(l)
del l
print(l)
```

```python
# deleting single element
# we have to give the index position
l = [1,2,3,4]
del l[2]
print(l)
```

```python
# remove
l = [1,2,3,4,5]
# to remove the element we just have to provide the value only
# no need to provide the value

l.remove(4)
print(l) 
```

```python
# pop
l = [1,2,3,4,5]
l.pop(1)
print(l)
```

```python
# if we don't pass the index value then last element is deleted by default
l = [1,2,3,4,5,6]

l.pop()
print(l)
```

```python
# clear
# it doesn't deletes the list but clear out the entire elements from the list
l = [1,3,4,5,6,7]
l.clear()
print(l)
```

## Operations on list

1. Arithmetic
2. Membership
3. Loop

```python

# Arithmetic
# we can merge two list by the plus operator
l = [1,2,3,4,5]
s= [6,7,8,9]
print(l + s)
```

```python
print(l*2)
```

```python
# Membership  operator
l = [1,2,3,4,5]
s = [6,7,8,9]
p = [[5,6],7,8,9]
print(s not in l)
print(l[-1] in p[0][0])
```

This is invalid because the "in" operator requires the right-hand side to be an iterable (such as a list, string, tuple, set, or dictionary).

```python
# Loops
l = [1,2,3,4,5]

for i in l:
    print(i)
```

## List functions

```python
# len/min/max/sorted
l = [1,23,4,5,6,8]
print(len(l))
print(min(l))
print(max(l))
print(sorted(l)) # not a permanent operation
print(sorted(l,reverse=True))
```

```python
# count
l = [1,2,3,4,1,4,6,3,3]
l.count(3)
```

```python
# index
l = [1,2,3,4,5]
l.index(2)
```

```python
# reverse
l = [1,2,4,5,6,7]
l.reverse()
# changes the list inplace
print(l)
```

```python
l = [1,2,4,5,6,7]

print(l.reverse())
```

Why does reverse() return None?

Many list methods that modify the list return None on purpose. This makes it clear that the method changes the existing list rather than creating a new one.

```python
# sort
# sort is permanent operation
l = [1,2,3,44,5,22,12]
l.sort()
print(l)
```

```python
# copy ---> shallow copy
l = [1,2,3,4,5]
print(id(l))
s = l.copy()
print(id(s))
print(s)
```
