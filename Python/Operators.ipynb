
# Operators In Python

```python
# Arithmetic Operators
# Basically all the mathematical operaots
print(5//2) # Floor Division or Integer Division
print(5+6)
print(5 % 2) # gives remainder
print(5 ** 3) # Power 
```

```python
# Relational Operator
print(4 < 5)
```

```python
# Logical Operator

print( 1 and 0)

print( 1 or 0)

print( not 1)
```

```python
# Bitwise Operator
# Specail operator which operates on binary values

print( 2 & 3) # Bitwise and operator

print( 2 | 3) # Bitwise or operator

print( 2 ^ 3) # Bitwise XOR operation

print(~3) # Bitwise not operator
```

```python
# Assignment Operator
# =
# a = 3
a = 3
a += 4

print(a)
```

```python
# Membership Operator

# in/ not in

print("D" in "Delhi")
print("O" not in "Hello World")
```

```python
print(1 in [2,3,4,5])
```

```python
# Program --> Find the sum of a 3 digit number entered by the user

num = int(input("Enter the number : "))
def sum_of_digit(num):
    res = 0
    while(num != 0):
        a = num % 10
        res = res + a
        num = num // 10
    return res

output = sum_of_digit(num)
print(f" The sum of the number {num} is = " , output)
```

```python
# Finding Minimum between 3 numbers
a = int(input("Enter the first number : "))
b = int(input("Enter the second number : "))
c = int(input("Enter the third number : "))

small = a
if small > b:
    small = b
elif small > c:
    small = c
else:
    small = a

print("Smallest Number is : ",small)
```

# Loops in Python


```python
# while loop

num = int(input("Enter the number = "))
i = 1
while i <= 10:
    print( num ,"*",i,"=",num*i)
    i = i + 1
    
```

```python

# else with while loop

x = 1
while x < 3:
    print(x)
    x = x + 1

else:
    print("limit crossed")
```

```python
# Guessing Game
import random

tar_num = random.randint(0,100)

attempt = 0
flag = False
while flag == False:
    try:
            
        guess = int(input("Enter your guess number"))
        attempt = attempt + 1
        if guess == tar_num:
            
            print(f"You guess the right number in {attempt} attempts")
            print(f"The random number to be guessed was {tar_num}")
            flag = True
        elif guess < tar_num : 
            print(f" Guess a bit higher than {guess} number")
        else:
            print(f"Guess a bit lower than {guess} number")
    except ValueError:
        print("Enter a valid integer")
```

```python
# for loops

for i in range(1,11):  # when using range function , second parameter doesn't get included
    
    print(i,end='\t')
```

```python
# to print the reverse order using range

for i in range(10,0,-1): # Step size is -1
    print(i,end="\t")
```
