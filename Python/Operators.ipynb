
# Operators In Python

```python
# Arithmetic Operators
# Basically all the mathematical operaots
print(5//2) # Floor Division or Integer Division
print(5+6)
print(5 % 2) # gives remainder
print(5 ** 3) # Power 
```

```python
# Relational Operator
print(4 < 5)
```

```python
# Logical Operator

print( 1 and 0)

print( 1 or 0)

print( not 1)
```

```python
# Bitwise Operator
# Specail operator which operates on binary values

print( 2 & 3) # Bitwise and operator

print( 2 | 3) # Bitwise or operator

print( 2 ^ 3) # Bitwise XOR operation

print(~3) # Bitwise not operator
```
