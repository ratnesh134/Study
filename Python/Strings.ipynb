```python
# Count the frequency of a particular character in a provided string
```

```python
s = input("Enter the string ")
char = input("Enter the character you want to know the frequency of ")
count = 0
for i in s:
    if i == char:
        count = count + 1
    else:
        continue

print(f"The frequency of {char} is {count}")
```

```python
# Palindrome String
```

```python
def palindrome(s : str):
    rev = s[::-1]
    if rev == s:
        print("Palindrom String")
    else:
        print("Not a palindrom String")

s = input("Enter the string to check ")
palindrome(s)
```

```python
name = "Ratnesh"

rev = name[::-1]

print(rev)
print(type(rev))
```

```python
# WAP to count the number of words in a string without split

s = input("Enter the string ")
count = 0
for i in s.strip():
    if i != " ":
        continue
    else:
        count = count + 1

print(f"The number of words in {s} is {count+1}")
```

```python
# WAP that convert an integer into a string
```

```python
number = int(input("Enter the number "))
digits = '0123456789'

result =''
while number != 0:
    result =  digits[number % 10] + result
    number = number // 10

print(result)
print(type(result))
```
