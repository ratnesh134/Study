# Strings

Strings are sequence of unicode characters.

Python Strings are immutable . Does not change over time.

We can delete the string as a whole but python string does not support partial deletion.

```python
# String Creation

s = str('hello')
print(s)
print(type(s))
```

```python
# Acessing the string contents

print(s[3])
```

```python
# Negative Indexing
s = "Hello World"
print(s[-2])
```

```python
print(s[::-1])
```

```python
# Arithematic Operations

print("Paris"+"Delhi")
```

```python
# Relational Operator
# All relational operator works in python string
'mumbai' > 'pune'

# Logic applied here is Python strings are compared lexiographically
```
