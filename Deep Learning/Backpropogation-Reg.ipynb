```python
def initalize_parameters(layer_dims):

    np.random.seed(3)
    parameters = {}
    L = len(layer_dims)

    for i in range(1,L):

        parameters['W' + str(i)] = np.ones((layer_dims[i-1],layer_dims[i])) * 0.1
        parameters['b' + str(i)] = np.zeros((layer_dims[i],1))

    return parameters
        
```

```python
initalize_parameters([2,2,1])
```
