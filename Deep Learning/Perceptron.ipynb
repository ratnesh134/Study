```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement.csv')
```

```python
print(df.head())
```

```python
df.shape
```

```python
sns.scatterplot(x='cgpa', y='resume_score', hue='placed', data=df)
plt.title('CGPA vs Resume Score')
plt.xlabel('CGPA')
plt.ylabel('Placement Exam Marks')
plt.show()
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X,y)
```

```python
# Weights
p.coef_
```

```python
# Bias
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values, clf=p, legend=2)
```
