## What is Lists ?

list is a data type where you can store multiple items under 1 name. list act like a dynamic arrays which means you can add more items on the fly.

## How lists are stored in memory ?

Referential Array where list contains the address where the value is stored.




To prove that list contains the address of the value-->




```python
a = 2

print(id(a))
```

```python
l = [1,2,3]

print(id(l))
print(id(l[0]))
print(id(l[1]))
print(id(l[2]))
print(id(1))
print(id(2))
print(id(3))
```

Benefits of storing address is that the list can be non-homogeneous
